```python
class Vuelo:
    def __init__(self, origen, destino, hora):
        self.origen = origen
        self.destino = destino
        self.hora = hora

    def leerdatos(self):
        self.origen = input("Ingrese el origen del vuelo: ")
        self.destino = input("Ingrese el destino del vuelo: ")
        self.hora = input("Ingrese la hora de salida: ")
        print("-----------------------------------")

    def mostrardatos(self):
        print("Origen: ", self.origen)
        print("Destino: ", self.destino)
        print("Hora: ", self.hora)

# Creando los 2 objetos (vuelo1, vuelo2)
vuelo1 = Vuelo("", "", "")
vuelo2 = Vuelo("", "", "")

# Procesando datos para vuelo1
print("DATOS DEL VUELO 1")
vuelo1.leerdatos()
vuelo1.mostrardatos()
print("*_*_*_*_*_*_*_*_*_*_*_*_*_*")

# Procesando datos para vuelo2
print("DATOS DEL VUELO 2")
vuelo2.leerdatos()
vuelo2.mostrardatos()
print("*_*_*_*_*_*_*_*_*_*_*_*_*_*")

print("Software desarrollado por Yahir Sifuentes NL = 58")
```
